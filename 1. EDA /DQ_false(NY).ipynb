```python
events2 = pd.read_csv('/content/drive/MyDrive/심화/events.csv')
```

```python
product_split = events2[events2['uri'].str.contains('product/')]
product_split = product_split.uri.str.split('/')
product_id = product_split.str.get(2)
events2['product_id']=product_id
```

```python
events2.sort_values(by = ['session_id', 'created_at'], ascending = True, inplace=True)

events2 = events2[['session_id','created_at','event_type','product_id','browser','traffic_source']]
events2['created_at'] = pd.to_datetime(events2['created_at'])
events2['date_diff'] = events2.groupby(['session_id'])['created_at'].diff().fillna(pd.Timedelta('0 days'))
events2['date_diff_seconds'] = events2['date_diff'].apply(lambda x: x.total_seconds())

events2['boolean_time_diff_30'] = events2['date_diff_seconds']>1800


events2_data=events2[events2['boolean_time_diff_30']==True]
true_session_id=events2_data['session_id']
```
